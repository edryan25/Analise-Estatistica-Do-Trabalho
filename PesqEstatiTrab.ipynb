```python
#Importando as bibliotecas
import pandas as pd
low_memory = False
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
#Abrindo o excel da base de dados
dados = r"C:\python\bdEstatisticaTrabalho.xlsx"

exc = pd.read_excel(dados)
exc1 = pd.read_excel(dados, 'Página1')
```

```python
exc1.columns
#Exibindo as colunas que a tabela possui
```

```python
avp=exc1.query('uf=="RJ"')  #nomendo a váriavel avp = RJ
avp.groupby('municipio')['admissoes_jan_2021'].max().sort_values().tail(10).plot(kind='barh',figsize=(10,10)) 
#Criando gráfico de barras horizontais que mostra as admissões das 10 cidades com indice maior do estado do Rio de Janeiro, no mês de janeiro.
```
